## Q-Q Plot

```python
import numpy as np
import pylab
import scipy.stats as stats

# Create a normal distribution with mean 0 and standard deviation 1
std_noraml = np.random.normal(
    loc=0, # mean
    scale=1, # standard deviation
    size=1000, # number of points
)

# Calculate percentile
for i in range(0, 101):
    print(i, np.percentile(std_noraml, i))
```

```python
# Check if the below distribuition is normal or not
measurements = np.random.normal(20, 5, 100)

stats.probplot(measurements, dist="norm", plot=pylab) # dist - norm creates and sets a normal distribution in x axis
pylab.show()
```

```python
measurements = np.random.normal(20, 5, 1000)

stats.probplot(measurements, dist="norm", plot=pylab) # dist - norm creates and sets a normal distribution in x axis
pylab.show()
```
