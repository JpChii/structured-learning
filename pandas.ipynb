```python
df_1 = pd.DataFrame({
    "city": ["Chennai", "Bangalore", 'Mumbai'],
    "humidity": [50, 45, 45]
})

df_2 = pd.DataFrame({
    "city": ["Chennai", "Bangalore", 'Hyderabad'],
    "temperature": [90, 80, 90]
})
```

```python
# Inner join, joins only if values are present in both df's for specified column
pd.merge(df_1, df_2, on="city")
```

We're merging based in city column, since hyderabad is present in one df alone, it's ignored on merge. To include hyderabad in merge, we need to use how="outer" param in merge.

```python
# Outer join
pd.merge(df_1, df_2, on="city", how="outer")
```
